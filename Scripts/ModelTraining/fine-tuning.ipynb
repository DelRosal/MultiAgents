```python
import supervision as sv
import ultralytics
import dotenv
import os
from roboflow import Roboflow
from IPython import display
display.clear_output()
ultralytics.checks()
```

### Set paths

```python
cwd = os.getcwd()
HOME = os.path.join(cwd)
env_path = os.path.join(HOME, '.env')
dataset_path = os.path.join(HOME, 'datasets', 'dump-truck-detection')
dotenv.load_dotenv(dotenv_path=env_path)
print(HOME)
```

### Download of the Dataset
Before you run this, check that you have created a **.env** file with a *ROBOFLOW_API_KEY=YOUR_TOKEN* variable

```python
ROBOFLOW_API_KEY = os.environ.get('ROBOFLOW_API_KEY')

rf = Roboflow(api_key=ROBOFLOW_API_KEY)

project = rf.workspace("tec-road7").project("materialbank")
dataset = project.version(5).download("yolov8", location=dataset_path)
```
